```python
# ===================================================================
# Part 1: 安装并配置 Miniconda
# ===================================================================
import os
import sys

# 检查 Conda 是否已经安装，避免重复执行
if not os.path.exists("/usr/local/bin/conda"):
    print("🚀 开始安装 Miniconda...")
    # 1. 下载 Miniconda 安装脚本
    !wget https://repo.anaconda.com/miniconda/Miniconda3-latest-Linux-x86_64.sh

    # 2. 以批处理(-b)和更新模式(-u)静默安装到 /usr/local 路径(-p)
    #    【修正点】添加了 -u 标志来处理 '/usr/local' 已存在的问题
    !bash Miniconda3-latest-Linux-x86_64.sh -b -u -p /usr/local

    # 3. 清理安装脚本
    !rm Miniconda3-latest-Linux-x86_64.sh
    print("✅ Miniconda 安装完成。")
else:
    print("✅ Miniconda 已安装，跳过安装步骤。")

# 4. 将 Conda 的 bin 目录添加到系统 PATH 环境变量中
conda_bin_path = "/usr/local/bin"
if conda_bin_path not in os.environ['PATH']:
    os.environ['PATH'] = f"{conda_bin_path}:{os.environ['PATH']}"

# ===================================================================
# Part 2: 创建并激活（伪激活）Conda 环境
# ===================================================================
env_name = "tf-gpu"
print(f"\n🔧 正在创建或检查 Conda 环境: {env_name}...")

# 检查环境是否已存在
if not os.path.exists(f"/usr/local/envs/{env_name}"):
    !conda create -n {env_name} python=3.7 cudatoolkit=10.0 cudnn=7.6 -c anaconda -c conda-forge -y
    print(f"✅ 环境 {env_name} 创建完成。")
else:
    print(f"✅ 环境 {env_name} 已存在，跳过创建步骤。")

# 2. 将 tf-gpu 环境的包路径添加到 Colab 的 Python 内核中
python_version = "3.7"
site_packages_path = f"/usr/local/envs/{env_name}/lib/python{python_version}/site-packages"
if site_packages_path not in sys.path:
    sys.path.append(site_packages_path)
    print(f"路径 {site_packages_path} 已添加至 sys.path。")

# ===================================================================
# Part 3: 在指定环境中安装 Python 包
# ===================================================================
print(f"\n📦 正在使用 pip 在 {env_name} 环境中安装/验证指定的包...")
pip_path = f"/usr/local/envs/{env_name}/bin/pip"
!{pip_path} install tensorflow-gpu==1.15.0 keras==2.2.4 opencv-python tqdm --index-url https://pypi.tuna.tsinghua.edu.cn/simple
!{pip_path} uninstall -y protobuf
!{pip_path} install protobuf==3.19.4 --index-url https://pypi.tuna.tsinghua.edu.cn/simple

print("\n🎉🎉🎉 环境设置完成！现在可以在新的单元格中运行您的项目代码。🎉🎉🎉")

# ===================================================================
# Part 4: 验证环境
# ===================================================================
print("\n🔍 开始验证环境...")
!/usr/local/envs/{env_name}/bin/python -c "import tensorflow as tf; print('TensorFlow 版本:', tf.__version__); print('GPU 可用状态:', tf.test.is_gpu_available())"
```
